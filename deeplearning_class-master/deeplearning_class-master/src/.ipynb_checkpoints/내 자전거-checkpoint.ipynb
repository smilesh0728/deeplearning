```python
# 딥러닝을 구동하는 데 필요한 케라스 함수를 불러옵니다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 필요한 라이브러리를 불러옵니다.
import numpy
import tensorflow as tf
import pandas as pd

# 실행할 때마다 같은 결과를 출력하기 위해 설정하는 부분입니다.
seed=0
numpy.random.seed(3)
tf.random.set_seed(3)
```

## 1. 데이터 읽어오기

```python
df = pd.read_csv("../dataset/bike_sharing_demand_train.csv")#, header=None)
print(df.head())
```

```python
print(df.info())
```

```python
df['year'] = df['datetime'].astype('datetime64').dt.year
df['month'] = df['datetime'].astype('datetime64').dt.month
df['day'] = df['datetime'].astype('datetime64').dt.day
```

```python
print(df.info())
```

### 2.데이터 전처리

```python
#  df.rename(columns={"A": "a", "B": "c"})
df = df.rename(columns={"count": "z-count"})
df = df.sort_index(axis=1)
df
```

```python
del (df['datetime'])
df
```

```python
X = df.values[:, 0:-1]
```

```python
Y= df.values[:, -1]
Y
```

```python
# 학습셋과 테스트셋을 나눔
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

```python
X_train
```

### 3. 모델 설계

```python
model = Sequential()
model.add(Dense(100, input_dim=13, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
# R 아니면 M이라 ->0,1로 나오게 하는 게 아니다, 그래서 그냥 숫자값이 나와야 하므로 넥티베이션? 그걸 지워버린다. 
model.add(Dense(1))
```

### 4. 모델 컴파일

```python
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics =['accuracy'])
# model.compile(loss = 'BCEE', optimizer = 'adam', metrics =['accuracy'])
```

### 5. 학습

```python
model.fit(X_train, Y_train, epochs = 200, batch_size = 4)
```
